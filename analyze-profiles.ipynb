```python
import json
import pandas as pd
```

```python
with open("profiles.json") as fp:
    profiles = json.load(fp)
```

```python
from collections import Counter

fields_counter = Counter()
for profile in profiles:
    fields = set(field.get("name") for field in profile["fields"])
    fields_counter.update(fields)


for key, count in fields_counter.items():
    print(f"{key}: {count}")
```

```python
import re
# as per recommendation from @freylis, compile once only
HTML_EXP = re.compile('<.*?>') 

def strip_tags(raw_html: str):
    return re.sub(HTML_EXP, "", raw_html)
```

```python
text = df.note.str.replace(HTML_EXP, "", regex=True).str.cat(sep=" ")
# text = strip_tags(text)
text[:1000]
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import numpy as np

def plot_cloud(wordcloud):
    plt.figure(figsize=(40, 30))
    plt.imshow(wordcloud) 
    plt.axis("off")


# circle mask
width, height = 1000, 1000
gap = 20
x, y = np.ogrid[:width, :height]
mask = (x - width//2) ** 2 + (y - height//2) ** 2 > (width//2-gap) ** 2
mask = 255 * mask.astype(int)
mask.shape


    
wordcloud = WordCloud(
    width=width,
    height=height,
    random_state=1,
    background_color='white',
    colormap='plasma',
    collocations=False,
    stopwords=STOPWORDS,
    mask=mask).generate(text)

plot_cloud(wordcloud)
```
